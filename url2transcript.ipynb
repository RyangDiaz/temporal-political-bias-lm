```python
from bs4 import BeautifulSoup
import requests
import time
import pandas as pd
import os
```

```python
def get_body(url):
    headers = {'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_2) AppleWebKit/601.3.9 (KHTML, like Gecko) Version/9.0.2 Safari/601.3.9'}
    response = requests.get(url, headers=headers)
    article_text = ''
    soup = BeautifulSoup(response.text, 'html.parser')
    print(response.text)
    paragraph = soup.find_all('p')
    for i in paragraph:
        a = i.get_text()
        if a != 'Advertisement' and a != 'Supported by' and a != 'Send any friend a story' and a != 'As a subscriber, you have 10 gift articles to give each month. Anyone can read what you share.' and not a.startswith("By"):
            article_text += a
            article_text += " "
    time.sleep(4)
    return article_text
```

```python
directory = '/content/drive/MyDrive/headlines/'


for filename in os.listdir(directory):
        if filename.endswith('.csv'):  # Check whether the file is a CSV
            file_path = os.path.join(directory, filename)  # Get the full path of the file
            data = pd.read_csv(file_path)  # Read the CSV file into a DataFrame
            print('Working on ' + filename + '...')

            # Check if 'web_url' column exists in the DataFrame
            if 'web_url' in data.columns:
                # Create a DataFrame to store the results
                results_df = pd.DataFrame(columns=['URL', 'Content'])
                #with open("/content/drive/MyDrive/Semester 6/5541 Team Folder/Project/nyt_transcripts/transcripts-" + filename, 'w', newline='') as file:
                csv_path = "/content/drive/MyDrive/Semester 6/5541 Team Folder/Project/nyt_transcripts/transcripts-" + filename
                if not os.path.exists(csv_path): # If we don't already have this month
                  transcripts = {'body': []}

                  # Iterate over each row in the 'web_url' column
                  i = 1
                  for url in data['web_url']:
                    i = i + 1
                    print('Working on url ' + str(i) + '/100...')

                    content = get_body(url)  # Get the content from the URL
                    transcripts['body'].append(content)

                  # Write the results DataFrame to the CSV file
                  results_df = pd.DataFrame(transcripts)
                  results_df.to_csv(csv_path, index=False)
            else:
                print(f"No 'web_url' column found in {filename}")
        else:
            continue
```
