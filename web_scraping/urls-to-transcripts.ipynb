```python
with open("/content/drive/MyDrive/urls-ltm.csv", 'r', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    #headers = next(reader)

    for row in reader:
        #print(row)


        url = "https://cnn.com" + row['url']

        response = requests.get(url)
        if response.status_code == 200:
            date = extract_date(row['url'])
            print("working on " + url)
            save_article(response, date)
        else:
            print(url + " not found")
```
